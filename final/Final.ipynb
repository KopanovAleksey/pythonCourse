```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
lst += ['kiborg'] * 10
lst += ['reptiloid'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data
```

```python
sl ={}
for i in set(lst):
  sl[i] = list()
print(sl)
```

```python
for i in set(lst):
  for j in lst:
    if i == j:
      sl[i].append(1)
    else:
      sl[i].append(0)
print(sl)
```
